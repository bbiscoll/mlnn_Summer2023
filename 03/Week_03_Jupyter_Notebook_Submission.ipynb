## 2.  Create a Linear Regression model (LIKE WE DID IN CLASS) to fit the data. *Use the example from Lesson 3 and DO NOT USE a library that calculates automatically*. We are expecting 5 coefficients to describe the linear model. 

## After creating the model (finding the coefficients), calculate a new column $y_p=\Sigma\beta_n\cdot x_n$ 

```python
#np.polyfit(x[0], y, 1)
# Coefficient 0 
m_numer_0 = np.sum(x.T[0]) * np.sum(y)/n - np.sum(x.T[0] * y) 
m_denom_0 = np.sum(x.T[0]) * np.sum(x.T[0])/n - np.sum(x.T[0] * x.T[0]) 
m_numer_0
m_denom_0
coeff_0 = m_numer_0 / m_denom_0
coeff_0

b_0 = (np.sum(x.T[0] * x.T[0]) * np.sum(y) - np.sum(x.T[0]) * np.sum(x.T[0] * y))/(n * np.sum(x.T[0]*x.T[0]) - np.sum(x.T[[0]]) * np.sum(x.T[[0]]))
coeff_0
#b_0
#np.polyfit(x.T[0], y, 1)
```

```python
# Coefficients
m_numer_1 = np.sum(x.T[1]) * np.sum(y)/n - np.sum(x.T[1] * y) 
m_denom_1 = np.sum(x.T[1]) * np.sum(x.T[1])/n - np.sum(x.T[1] * x.T[1]) 

m_numer_2 = np.sum(x.T[2]) * np.sum(y)/n - np.sum(x.T[2] * y) 
m_denom_2 = np.sum(x.T[2]) * np.sum(x.T[2])/n - np.sum(x.T[2] * x.T[2]) 

m_numer_3 = np.sum(x.T[3]) * np.sum(y)/n - np.sum(x.T[3] * y) 
m_denom_3 = np.sum(x.T[3]) * np.sum(x.T[3])/n - np.sum(x.T[3] * x.T[3])

coeff_1 = m_numer_1 / m_denom_1
coeff_2 = m_numer_2 / m_denom_2
coeff_3 = m_numer_3 / m_denom_3

# b Values

b_1 = (np.sum(x.T[1] * x.T[1]) * np.sum(y) - np.sum(x.T[1]) * np.sum(x.T[1] * y))/(n * np.sum(x.T[1]*x.T[1]) - np.sum(x.T[[1]]) * np.sum(x.T[[1]]))
b_2 = (np.sum(x.T[2] * x.T[2]) * np.sum(y) - np.sum(x.T[2]) * np.sum(x.T[2] * y))/(n * np.sum(x.T[2]*x.T[2]) - np.sum(x.T[[2]]) * np.sum(x.T[[2]]))
b_3 = (np.sum(x.T[3] * x.T[3]) * np.sum(y) - np.sum(x.T[3]) * np.sum(x.T[3] * y))/(n * np.sum(x.T[3]*x.T[3]) - np.sum(x.T[[3]]) * np.sum(x.T[[3]]))

# b values are constant, so they are summed
b_final = b_0 + b_1 + b_2 + b_3 
b_final
```

```python
# y = m1 * x + b
# y = (m1 * x)+(m2 * x)...+(m5 * x) + b
# y = x *(m1+m2+m3+m4) + b
# y - b = m * x 
# ( y - b ) / m = x 
# ( y - b ) / m = x 
y_p0 = (coeff_0+coeff_1+coeff_2+coeff_3)*(x.T[0]) + b_final
y_p1 = (coeff_0+coeff_1+coeff_2+coeff_3)*(x.T[1]) + b_final
y_p2 = (coeff_0+coeff_1+coeff_2+coeff_3)*(x.T[2]) + b_final
y_p3 = (coeff_0+coeff_1+coeff_2+coeff_3)*(x.T[3]) + b_final

y_p1
x.T[0]
# plt.scatter(x.T[0],y_p1)
```

```python
# y = m1 * x + b
# y = (m1 * x)+(m2 * x)...+(m5 * x) + b
# y = x *(m1+m2+m3+m4) + b
# y - b = m * x 
# ( y - b ) / m = x 
# ( y - b ) / m = x 
y_p4 = (coeff_0)*(x.T[0]) + (coeff_1)*(x.T[1]) + (coeff_2)*(x.T[2]) + (coeff_3)*(x.T[3]) + b_final

y_p4
# x.T[0]
# plt.scatter(x.T[0],y_p1)
```
